2. No obvious relationship between latitude and humidity. Requires further analysis, perhaps introduction of longitude.

2. No obvious relationship between latitude and wind speed. Requires further analysis, perhaps introduction of longitude.

```python
#Importing Libraries
import matplotlib.pyplot as plt
import seaborn
import requests
import random
import pandas as pd
import numpy as np
from citipy import citipy
from pprint import pprint
from config import OWM_API_key
```

```python
lat=[]
long=[]
while len(lat)<=1450:
    lat.append(random.uniform(-90.00, 90.00)) 
while len(long)<=1450:
    long.append(random.uniform(-180.00, 180.00))
print(len(lat), len(long))
```

```python
citylist=[]
for lt, ln in zip(lat,long):
    city=citipy.nearest_city(lt,ln)
    cityname=city.city_name
    if cityname not in citylist:
        citylist.append(cityname)
print(len(citylist))
```

```python
url = "http://api.openweathermap.org/data/2.5/weather"
query_params = {'appid': OWM_API_key,'units': 'Imperial'}
temp=[]
hum=[]
cloud=[]
wind=[]
Latitude=[]
Longitude=[]
check=0
for y in citylist:
    query_params['q'] = y

    try:
        cityurl=requests.get(url, params=query_params)
        response = cityurl.json()
        temp.append(response['main']['temp_max'])
        hum.append(response['main']['humidity'])
        cloud.append(response['clouds']['all'])
        wind.append(response['wind']['speed'])
        Latitude.append(response['coord']['lat'])
        Longitude.append(response['coord']['lon'])
        print({cityurl.url})
        print(check,y,temp[check],hum[check],cloud[check],wind[check],Latitude[check],Longitude[check])
    except KeyError:
        temp.append('')
        hum.append('')
        cloud.append('')
        wind.append('')
        Latitude.append('')
        Longitude.append('')
    check=check+1
```

```python
print('Temp:'+str(len(temp)))
print('hum:'+str(len(hum)))
print('Cloud:'+str(len(cloud)))
print('Wind:'+str(len(wind)))
print('Lat:'+str(len(Latitude)))
print('Long:'+str(len(Longitude)))
```

```python
weather={'city':citylist,'lat':Latitude,'long':Longitude,'temp':temp,'humidity':hum,'cloud':cloud,'wind':wind}
weather_df=pd.DataFrame(weather)
weather_df.head()
# weather_df.dtypes
```

```python

weather_df = weather_df.drop(weather_df[(weather_df.temp == '')].index)
weather_df['cloud']=weather_df['cloud'].astype(int)
weather_df['humidity']=weather_df['humidity'].astype(int)
weather_df['temp']=weather_df['temp'].astype(float)
weather_df['wind']=weather_df['wind'].astype(float)
weather_df['lat']=weather_df['lat'].astype(float)
weather_df['long']=weather_df['long'].astype(float)

# weather_df.dtypes
weather_df.describe()
```

```python
# weather_df.plot(kind='scatter',x=weather_df['lat'],y=weather_df['temp'])
fig=plt.figure(figsize=(10,7))
plt.scatter(weather_df['lat'],weather_df['temp'], marker='o')
plt.xlim(-80,100)
plt.ylim(-100,150)
plt.title('City Latitude vs. Max Temperature')
plt.ylabel('Max Temperature (F)')
plt.xlabel('Latitude')
plt.grid()
plt.savefig("City_Latitude_vs_Max_Temperature.png", dpi = 100)
plt.show()
```

```python
fig=plt.figure(figsize=(10,7))
plt.scatter(weather_df['lat'],weather_df['humidity'])
plt.xlim(-80,100)
plt.ylim(-20,120)
plt.title('City Latitude vs. Humidity')
plt.ylabel('Humidity')
plt.xlabel('Latitude')
plt.grid()
plt.savefig("City_Latitude_vs_Humidity.png", dpi = 100)
plt.show()
```

```python
fig=plt.figure(figsize=(10,7))
plt.scatter(weather_df['lat'],weather_df['cloud'])
plt.xlim(-80,100)
plt.ylim(-20,120)
plt.title('City Latitude vs. Cloudiness Plot')
plt.ylabel('Cloudiness')
plt.xlabel('Latitude')
plt.grid()
plt.savefig("City_Latitude_vs_Cloudiness.png", dpi = 100)
plt.show()
```

```python
fig=plt.figure(figsize=(10,7))
plt.scatter(weather_df['lat'],weather_df['wind'])
plt.xlim(-80,100)
plt.ylim(-5,40)
plt.title('City Latitude vs. Wind Speed Plot')
plt.ylabel('Wind Speed')
plt.xlabel('Latitude')
plt.grid()
plt.savefig("City_Latitude_vs_Wind_Speed.png", dpi = 100)
plt.show()
```
